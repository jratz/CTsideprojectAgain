```python
import string
from time import time


file = open('data/small_vocab_en', 'r')
text = file.read()

a_z = string.ascii_lowercase
```

```python
i_count = 0
for char in text:
    if char in a_z[17]:
        i_count += 1
i_count
```

```python
characters = list(a_z)
values = [0] * len(characters)
for char in text:
    if char in a_z:
        values[characters.index(char)] += 1
values
```
